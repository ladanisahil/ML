```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("bank_transactions.csv")
data
```

```python
data['TransactionDate']= pd.to_datetime(data['TransactionDate'],format='%d-%m-%y')
```

```python
data.dropna(inplace=True)
```

```python
numeric_feature = data[['CustAccountBalance','TransactionAmount (INR)']]
```

```python
Scaler = StandardScaler()
Scaled_feature = Scaler.fit_transform(numeric_feature)
```

```python
wcss =[]
for i in  range(1,11):
    k_means= KMeans(n_clusters = i, init='k-means++',random_state=42)
    k_means.fit(Scaled_feature)
    wcss.append(KMeans.inertia_)


plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss,Markdown='o',)
plt.xlabel('Numberof cluster(k)')
plt.ylabel('wcss')
plt.title("eblow method to find optional k")
plt.show()
 
```
